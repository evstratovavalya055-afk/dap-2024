### Задание 5: Выведите распределение переменной Pclass и это же распределение для мужчин/женщин. Сколько было мужчин 2-го класса?


````python
# Распределение Pclass
print("Распределение по классам обслуживания:")
print(df['Pclass'].value_counts().sort_index())
print()

# Распределение по классам для мужчин и женщин
print("Распределение по классам и полу:")
class_sex = pd.crosstab(df['Pclass'], df['Sex'])
print(class_sex)
print()

# Мужчины 2-го класса
men_class2 = len(df[(df['Pclass'] == 2) & (df['Sex'] == 'male')])
print(f"Мужчин 2-го класса: {men_class2}")
````

### Задание 7: Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?


````python
# Сравнение выживаемости по возрасту
young = df[df['Age'] < 30]
old = df[df['Age'] > 60]

young_survival = young['Survived'].mean()
old_survival = old['Survived'].mean()

print(f"Количество людей младше 30: {len(young)}")
print(f"Доля выживших младше 30: {young_survival:.2%}")
print()
print(f"Количество людей старше 60: {len(old)}")
print(f"Доля выживших старше 60: {old_survival:.2%}")
print()
if young_survival > old_survival:
    print("Правда: люди моложе 30 лет выживали чаще, чем люди старше 60 лет")
else:
    print("Неправда: люди моложе 30 лет выживали реже")
````

### Задание 8: Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?


````python
# Сравнение выживаемости по полу
women_survival = df[df['Sex'] == 'female']['Survived'].mean()
men_survival = df[df['Sex'] == 'male']['Survived'].mean()

print(f"Количество женщин: {len(df[df['Sex'] == 'female'])}")
print(f"Доля выживших женщин: {women_survival:.2%}")
print()
print(f"Количество мужчин: {len(df[df['Sex'] == 'male'])}")
print(f"Доля выживших мужчин: {men_survival:.2%}")
print()
if women_survival > men_survival:
    print("Правда: женщины выживали чаще мужчин")
else:
    print("Неправда: женщины выживали реже")
````

### Задание 9: Найдите самое популярное имя среди пассажиров Титаника мужского пола


````python
# Поиск самого популярного мужского имени
men = df[df['Sex'] == 'male'].copy()

# Извлекаем имя (часть после титула)
def extract_first_name(full_name):
    try:
        parts = full_name.split(',')
        if len(parts) > 1:
            title_and_name = parts[1].strip()
            if '.' in title_and_name:
                name_part = title_and_name.split('.')[1].strip()
                return name_part.split()[0] if name_part else 'Unknown'
        return 'Unknown'
    except:
        return 'Unknown'

men['FirstName'] = men['Name'].apply(extract_first_name)
popular_names = men['FirstName'].value_counts()

print("Самые популярные мужские имена:")
print(popular_names.head(10))
print()
print(f"Самое популярное мужское имя: {popular_names.index[0]} (встречается {popular_names.iloc[0]} раз)")
````

### Задание 10: Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания?


````python
# Средний возраст по классам и полу
avg_age = df.groupby(['Pclass', 'Sex'])['Age'].mean().unstack()
print("Средний возраст по классам и полу:")
print(avg_age)
print()

# Проверка утверждений
print("Проверка утверждений:")
print(f"1. Мужчины 1-го класса старше 40 лет: {avg_age.loc[1, 'male'] > 40}")
print(f"2. Женщины 1-го класса старше 40 лет: {avg_age.loc[1, 'female'] > 40}")
print(f"3. Мужчины всех классов старше женщин того же класса: {all(avg_age.loc[cls, 'male'] > avg_age.loc[cls, 'female'] for cls in [1,2,3])}")
print(f"4. Люди в 1 классе старше, чем во 2-ом, а те старше 3-го: {(avg_age.loc[1, 'male'] + avg_age.loc[1, 'female']) / 2 > (avg_age.loc[2, 'male'] + avg_age.loc[2, 'female']) / 2 > (avg_age.loc[3, 'male'] + avg_age.loc[3, 'female']) / 2}")
````

### Задание 11: Сравните возраст у спасенных и у погибших пассажиров. Средний возраст погибших выше, верно?


````python
# Сравнение возраста выживших и погибших
survived_age = df[df['Survived'] == 1]['Age'].mean()
died_age = df[df['Survived'] == 0]['Age'].mean()

print(f"Средний возраст выживших: {survived_age:.2f} лет")
print(f"Средний возраст погибших: {died_age:.2f} лет")
print()
if died_age > survived_age:
    print("Верно: средний возраст погибших выше")
else:
    print("Неверно: средний возраст погибших ниже или равен")
````

### 1. Что такое медиана? Рассчитать медиану для предложенного ряда чисел.

**Медиана** - это значение, которое делит упорядоченный ряд данных пополам (50% значений меньше медианы, 50% - больше).

```python
# Пример расчета медианы
numbers = [5, 2, 8, 1, 9, 3, 7]
sorted_numbers = sorted(numbers)
median = np.median(numbers)
print(f"Ряд: {numbers}")
print(f"Медиана: {median}")
```

---

### 2. Что такое дисперсия? Рассчитать дисперсию для предложенного ряда чисел.

**Дисперсия** - мера разброса данных относительно среднего значения.

```python
numbers = [5, 2, 8, 1, 9, 3, 7]
variance = np.var(numbers)
print(f"Дисперсия: {variance:.2f}")
```

std_dev = np.std(numbers)
