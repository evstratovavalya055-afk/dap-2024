### 1. Типы ячеек в Jupyter Notebook

В Jupyter Notebook существует **3 типа ячеек**:

- **Code (код)** — для выполнения программного кода на Python
- **Markdown** — для форматированного текста, заголовков, списков, изображений
- **Raw (сырой)** — текст без форматирования, не обрабатывается

#### Очистка данных (Data Cleaning)

Это процесс выявления и исправления ошибок, пропусков и несоответствий в данных.

**Основные действия:**

- **Обработка пропущенных значений:**
  ```python
  df.dropna()           # удаление строк с NaN
  df.fillna(value)      # заполнение пропусков
  ```

- **Удаление дубликатов:** `df.drop_duplicates()`

- **Исправление типов данных:** `df['column'] = df['column'].astype('int')`

- **Обработка выбросов (аномалий):**
  ```python
  # Метод IQR (межквартильный размах)
  Q1 = df['col'].quantile(0.25)
  Q3 = df['col'].quantile(0.75)
  IQR = Q3 - Q1
  df = df[(df['col'] >= Q1 - 1.5*IQR) & (df['col'] <= Q3 + 1.5*IQR)]
  ```

**Почему это важно?** Качество данных напрямую влияет на точность анализа и выводов.

### 7. Типы данных (подробно о типе **Категориальные данные**)

**Основные типы данных в анализе:**

| Тип | Описание | Пример |
|-----|----------|--------|
| Числовые | Количественные значения | 25, 3.14, -10 |
| **Категориальные** | Метки/категории | "красный", "Москва" |
| Временные | Даты и время | 2024-01-01 |
| Текстовые | Неструктурированный текст | отзывы, описания |

#### Категориальные данные (Categorical Data)

Это данные, которые принимают ограниченное количество уникальных значений (категорий).

**Подтипы:**

- **Номинативные (Nominal):** нет порядка
  - Цвет: красный, синий, зеленый
  - Город: Москва, СПб, Казань

- **Порядковые (Ordinal):** есть иерархия
  - Оценки: плохо, удовлетворительно, хорошо, отлично
  - Уровень образования: школа, бакалавр, магистр

**Кодирование категориальных данных:**

```python
# One-Hot Encoding
pd.get_dummies(df['color'])

# Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['color_encoded'] = le.fit_transform(df['color'])
```
