### Задача 1: Two Sum

**Ссылка:** [LeetCode 1. Two Sum](https://leetcode.com/problems/two-sum/)

**Условие:** Дан массив целых чисел `nums` и целое число `target`. Нужно вернуть индексы двух чисел, сумма которых равна `target`.

**Решение:**

```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        n = len(nums)
        for i in range(n):
            for j in range(i + 1, n):
                if nums[i] + nums[j] == target:
                    return [i, j]
        return []
```

**Алгоритм:** Перебор всех пар элементов.

### Задача 2: Palindrome Number

**Ссылка:** [LeetCode 9. Palindrome Number](https://leetcode.com/problems/palindrome-number/)

**Условие:** Определить, является ли целое число палиндромом (читается одинаково слева направо и справа налево).

**Решение:**

```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        str_x = str(x)
        if str_x == str_x[::-1]:
            return True
        else:
            return False
```

**Алгоритм:** Преобразуем число в строку и сравниваем с перевернутой версией.

### Задача 3: Roman to Integer

**Ссылка:** [LeetCode 13. Roman to Integer](https://leetcode.com/problems/roman-to-integer/)

**Условие:** Преобразовать римское число в целое.

**Решение:**

```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        d = {'I':1, 'V':5, 'X':10, 'L':50, 'C':100, 'D':500, 'M':1000}
        return sum(-d[s[i]] if i+1 < len(s) and d[s[i]] < d[s[i+1]] else d[s[i]] for i in range(len(s)))
```

**Алгоритм:** Если текущий символ меньше следующего, вычитаем его, иначе прибавляем.

### Задача 4: Longest Common Prefix

**Ссылка:** [LeetCode 14. Longest Common Prefix](https://leetcode.com/problems/longest-common-prefix/)

**Условие:** Найти самый длинный общий префикс среди массива строк.

**Решение:**

```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """
        if not strs:
            return ""
        s1 = min(strs)
        s2 = max(strs)
        for i, ch in enumerate(s1):
            if i >= len(s2) or ch != s2[i]:
                return s1[:i]
        return s1
```

**Алгоритм:** Используем лексикографический минимум и максимум.

### Задача 5: Merge Two Sorted Lists

**Ссылка:** [LeetCode 21. Merge Two Sorted Lists](https://leetcode.com/problems/merge-two-sorted-lists/)

**Условие:** Слить два отсортированных связных списка в один отсортированный список.

**Решение:**

```python
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        if not list1 or not list2:
            return list1 or list2
        if list1.val <= list2.val:
            list1.next = self.mergeTwoLists(list1.next, list2)
            return list1
        else:
            list2.next = self.mergeTwoLists(list1, list2.next)
            return list2
```

**Алгоритм:** Рекурсивное слияние.

### Задача 6: Longest Substring Without Repeating Characters

**Ссылка:** [LeetCode 3. Longest Substring Without Repeating Characters](https://leetcode.com/problems/longest-substring-without-repeating-characters/)

**Условие:** Найти длину самой длинной подстроки без повторяющихся символов.

**Решение:**

```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """
        d = {}
        left = ans = 0
        for right, c in enumerate(s):
            if c in d and d[c] >= left:
                left = d[c] + 1
            ans = max(ans, right - left + 1)
            d[c] = right
        return ans
```

**Алгоритм:** Скользящее окно (sliding window). Временная сложность: O(n), память: O(min(n, алфавит)).

### 1. Что такое динамическая типизация и чем она отличается от статической?

**Динамическая типизация** - тип переменной определяется во время выполнения программы, а не при компиляции.

| Динамическая (Python) | Статическая (C++, Java) |
|----------------------|------------------------|
| Тип проверяется во время выполнения | Тип проверяется во время компиляции |
| Переменная может менять тип | Тип переменной фиксирован |
| `x = 5; x = "hello"` - работает | `int x = 5; x = "hello"` - ошибка |
| Более гибкий, но медленнее | Более строгий, но быстрее |

### 2. Чем отличается цикл for в Python от цикла for в C++?

```python
# Python - итерация по последовательности
for item in [1, 2, 3]:
    print(item)
```

```cpp
// C++ - итерация с счетчиком
for (int i = 0; i < 3; i++) {
    cout << i << endl;
}
```

**Основные отличия:**
- Python: `for элемент in коллекция` (foreach-подобный)
- C++: `for (инициализация; условие; инкремент)`
- Python имеет `range()` для генерации чисел
- C++ требует ручного управления счетчиком

### 3. Для чего предназначены операторы continue и break?

- **`break`** - полностью прерывает выполнение цикла
- **`continue`** - пропускает текущую итерацию и переходит к следующей

```python
for i in range(5):
    if i == 2:
        continue  # пропускаем 2
    if i == 4:
        break     # выходим на 4
    print(i)      # вывод: 0, 1, 3
```

### 10. Чем staticmethod отличается от classmethod?

| `@staticmethod` | `@classmethod` |
|-----------------|----------------|
| Не получает специальных параметров | Получает `cls` (ссылку на класс) |
| Не может изменять состояние класса | Может изменять состояние класса |
| Как обычная функция внутри класса | Имеет доступ к атрибутам класса |

```python
class Example:
    @staticmethod
    def add(a, b):
        return a + b
    
    @classmethod
    def create(cls):
        return cls()
```
