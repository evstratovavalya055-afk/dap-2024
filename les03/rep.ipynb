### Задание 1: Создать массив 10x10 со случайными значениями, найти минимум и максимум

```python
arr = np.random.random((10, 10))
print("Массив 10x10:")
print(arr)
print(f"\nМинимум: {np.min(arr)}")
print(f"Максимум: {np.max(arr)}")
```

### Задание 2: Создать 8x8 матрицу и заполнить её единицами и нулями в шахматном порядке

```python
chess = np.zeros((8, 8), dtype=int)
chess[1::2, ::2] = 1
chess[::2, 1::2] = 1
print("Матрица в шахматном порядке:")
print(chess)
```

### Задание 3: Дан массив из 11 элементов, поменять знак у элементов, значения которых между 3 и 8

```python
z = np.arange(11)
print("Исходный массив:", z)
z[(z > 3) & (z < 8)] *= -1
print("После изменения знака:", z)
```

### Задание 4: Найти среднее значение строки матрицы и вычесть его из каждого элемента строки

```python
z = np.random.rand(5, 10)
print("Исходная матрица (5x10):")
print(z)
row_means = z.mean(axis=1, keepdims=True)
z_centered = z - row_means
print(f"\nСредние значения строк: {row_means.flatten()}")
print("\nМатрица после вычитания среднего:")
print(z_centered)
print(f"\nПроверка (сумма строк должна быть близка к 0): {z_centered.sum(axis=1)}")
```

### Задание 5: Поменяйте 1 и 2 строки матрицы

### Задание 6: Дан массив 10x2 (точки в декартовой системе), преобразовать в полярную

```python
z = np.random.random((10, 2))
print("Декартовы координаты (x, y):")
print(z)

x = z[:, 0]
y = z[:, 1]
r = np.sqrt(x**2 + y**2)
phi = np.arctan2(y, x)
polar = np.column_stack((r, phi))

print("\nПолярные координаты (r, φ):")
print(polar)
```

### Задание 7: Заменить максимальный элемент массива на ноль

```python
z = np.random.random(10)
print("Исходный массив:", z)
max_idx = np.argmax(z)
print(f"Максимальный элемент: {z[max_idx]} на индексе {max_idx}")
z[max_idx] = 0
print("После замены максимума на 0:", z)
```

### Задание 8: Добавить 1 к каждому элементу с индексом из другого массива (только 1 раз)

```python
z = np.ones(10)
i = np.random.randint(0, len(z), 20)
print("Исходный массив:", z)
print(f"Индексы для добавления (с повторами): {i}")

unique_indices = np.unique(i)
print(f"Уникальные индексы: {unique_indices}")
z[unique_indices] += 1
print("Результат:", z)
```

### Задание 9: Построить новый вектор с тремя нулями между каждым значением

```python
z = np.array([1, 2, 3, 4, 5])
print(f"Исходный вектор: {z}")

result = np.zeros(len(z) * 4 - 3, dtype=int)
result[::4] = z
print(f"Результат (с тремя нулями между значениями): {result}")
```

### Задание 10: Найти ближайшее к заданному значению число в заданном массиве

```python
z = np.arange(100)
v = np.random.uniform(0, 100)
print(f"Массив: от 0 до 99")
print(f"Заданное значение: {v:.4f}")

idx = np.argmin(np.abs(z - v))
print(f"Ближайшее число: {z[idx]}")
print(f"Расстояние: {abs(z[idx] - v):.4f}")
```

### 1. Какой тип (dtype) будут иметь элементы массива?

| Выражение | Тип (dtype) | Пояснение |
|-----------|-------------|------------|
| `np.array([1, 4, 2, 5, 3])` | `int64` или `int32` | Целые числа |
| `np.array([1., 4, 2, 5, 3])` | `float64` | Наличие точки → float |
| `np.array([1, 4, 2, 5, 3], dtype='float128')` | `float128` | Явно указан тип |
| `np.array([1., 4, 2, 5, 3], dtype='complex128')` | `complex128` | Явно complex |
| `np.array([1, 4, 2, 5, 3], dtype='float64')` | `float64` | Явно float64 |
| `np.array([1, 4, 2, 5, 3], dtype='bool')` | `bool` | 0=False, остальные True |
| `np.array([1, 4, 2, 5, '3'])` | `str` | Смешанные → строка |
| `np.array([1, -4, 2, 0, 1j])` | `complex128` | Наличие комплексного числа |

---

### 2. Что будет выведено в консоль и почему?

| Выражение | Результат | Пояснение |
|-----------|-----------|------------|
| `np.array([1, 4, 2, 5, 3], dtype='str')[1] + '1'` | `'41'` | Конкатенация строк |
| `np.array([1, -4, 2, 0, 3], dtype='bool')[3] or np.array([1, -4, 2, 0, 3])[1]` | `True` | 0=False, -4=True |
| `np.array([1, -4, 2, 0, 3], dtype='bool')[3] and np.array([1, -4, 2, 0, 3])[1]` | `False` | False and True = False |
| `np.array([1, -4, 2, 0, True], dtype='int64')[-1]` | `1` | True → 1 |
| `np.array([1, 4, 2, 5, '3'])[-1] * 3` | `'333'` | Строка повторяется 3 раза |

---

### 3. Каков результат операции?

| Операция | Результат |
|----------|-----------|
| `np.ones((3, 5), dtype=float)` | Матрица 3x5 из единиц |
| `np.full((3, 5), 3.14)` | Матрица 3x5 заполненная 3.14 |
| `np.linspace(0, 1, 5)` | `[0., 0.25, 0.5, 0.75, 1.]` |
| `np.diag((1,3,5))` | Диагональная матрица 3x3 |
| `np.random.random((3, 3))` | Матрица 3x3 со случайными числами |
| `np.eye(3)` | Единичная матрица 3x3 |
| `np.empty(3)` | Массив из 3 элементов (мусорные значения) |

### 4. В чем преимущество массивов NumPy по сравнению с обычными списками?

**Преимущества NumPy:**
1. Скорость работы - операции в 10-100 раз быстрее
2. Экономия памяти - однородные типы хранятся компактно
3. Векторизованные операции - без явных циклов
4. Удобные математические функции

**Причина:** реализация на C/Fortran, непрерывное хранение в памяти, отсутствие Python overhead.
