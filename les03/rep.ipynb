### 1. Какой тип (dtype) будут иметь элементы массива?

| Выражение | Тип (dtype) | Пояснение |
|-----------|-------------|------------|
| `np.array([1, 4, 2, 5, 3])` | `int64` или `int32` | Целые числа |
| `np.array([1., 4, 2, 5, 3])` | `float64` | Наличие точки → float |
| `np.array([1, 4, 2, 5, 3], dtype='float128')` | `float128` | Явно указан тип |
| `np.array([1., 4, 2, 5, 3], dtype='complex128')` | `complex128` | Явно complex |
| `np.array([1, 4, 2, 5, 3], dtype='float64')` | `float64` | Явно float64 |
| `np.array([1, 4, 2, 5, 3], dtype=int64)` | Ошибка | `int64` должна быть строкой `'int64'` |
| `np.array([1, 4, 2, 5, 3], dtype='bool')` | `bool` | 0=False, остальные True |
| `np.array([1, 4, 2, 5, 3], dtype='str')` | `str` (или `U<...`) | Строковый тип |
| `np.array([1, 4, 2, 5, '3'])` | `str` (или `U<...`) | Смешанные → строка |
| `np.array([1, -4, 2, 0, 1j])` | `complex128` | Наличие комплексного числа |

### 2. Что будет выведено в консоль и почему?

| Выражение | Результат | Пояснение |
|-----------|-----------|------------|
| `np.array([1, 4, 2, 5, 3], dtype='str')[1] + '1'` | `'41'` | Строковое сложение (конкатенация) |
| `np.array([1, -4, 2, 0, 3], dtype='bool')[3] or np.array([1, -4, 2, 0, 3])[1]` | `True` | `0` = False, `-4` = True, False or True = True |
| `np.array([1, -4, 2, 0, 3], dtype='bool')[3] and np.array([1, -4, 2, 0, 3])[1]` | `False` | `0` = False, `-4` = True, False and True = False |
| `np.array([1, -4, 2, 0, True], dtype='int64')[-1]` | `1` | True преобразуется в 1 |
| `np.array([1, -4, 2, 0, 1j], dtype='int64')[1]` | Ошибка | Нельзя преобразовать complex в int |
| `np.array([1, -4, 2, 0, 1j], dtype='str')[-1]` | `'1j'` | Комплексное число в строку |
| `np.array([1, 4, 2, 5, '3'])[-1] * 3` | `'333'` | Строка `'3'` повторяется 3 раза |

### 3. Каков результат операции?

| Операция | Результат |
|----------|-----------|
| `np.ones((3, 5), dtype=float)` | Матрица 3x5 из единиц (float) |
| `np.full((3, 5), 3.14)` | Матрица 3x5 заполненная 3.14 |

| `np.empty(3)` | Массив из 3 элементов (с мусорными значениями) |

### 4. Дан двумерный массив X, что будет выведено в консоль?

```python
X = np.array([[3, 5, 2, 4],
              [7, 6, 8, 8],
              [1, 6, 7, 7]])
```

| Выражение | Результат | Пояснение |
|-----------|-----------|------------|
| `X[0,0]` | `3` | 1-я строка, 1-й столбец |
| `X[-1,-1]` | `7` | последняя строка, последний столбец |
| `X[0:1,-1]` | `[4]` | 1-я строка, последний столбец |
| `X[0:2,-1]` | `[4, 8]` | первые 2 строки, последний столбец |
| `X[0:2,0:2]` | `[[3,5],[7,6]]` | подматрица 2x2 |
| `X[1,1:3]` | `[6, 8]` | 2-я строка, столбцы 1-2 |
| `X[1,:]` | `[7, 6, 8, 8]` | вся 2-я строка |
| `X[:,:]` | вся матрица | все строки и столбцы |
| `X[:]` | вся матрица | все строки |
| `X[::2,::2]` | `[[3,2],[1,7]]` | шаг 2 по строкам и столбцам |
| `X[0:3:2,0:3:2]` | `[[3,2],[1,7]]` | то же самое |
| `X[::-1,::-1]` | `[[7,7,6,1],[8,8,6,7],[4,2,5,3]]` | переворот по обеим осям |

### 5. Дан двумерный массив grid, в чем различие между операциями?

```python
grid = np.arange(16).reshape((4, 4))
```

- `np.split(grid, [2])` - **Разделение по строкам** (ось 0 по умолчанию). Получаем два массива: строки 0-1 и строки 2-3.

- `np.split(grid, [2], axis=1)` - **Разделение по столбцам**. Получаем два массива: столбцы 0-1 и столбцы 2-3.

### 7. Как в Colab измерить время выполнения кода?

**Способы измерения времени в Colab:**

```python
# Способ 1: Магическая команда %%time
%%time
# ваш код
```

### 8. Дан массив X, каковы результаты операций?

```python
M = np.array([[3, 5, 2, 4],
              [7, 6, 8, 8],
              [1, 6, 7, 7]])
```

| Выражение | Результат | Пояснение |
|-----------|-----------|------------|
| `np.sum(M)` | `57` | Сумма всех элементов |

### 10. В чем преимущество массивов NumPy по сравнению с обычными списками? Чем обусловлено это преимущество?

**Преимущества NumPy:**

1. **Скорость работы** - операции выполняются в 10-100 раз быстрее
2. **Экономия памяти** - элементы одного типа хранятся компактно
3. **Векторизованные операции** - без явных циклов
4. **Удобные математические функции** - линейная алгебра, статистика, FFT
5. **Broadcasting** - операции с массивами разной формы

**Причина преимущества:**

- **Непрерывное хранение в памяти** (C-style layout) - кэш-эффективность
- **Однородность типов** - элементы занимают фиксированный размер
- **Реализация на C/Fortran** - операции выполняются в скомпилированном коде
- **Отсутствие Python overhead** - нет проверок типов на каждом шаге

```python
# Пример сравнения скорости
lst = list(range(1000000))
arr = np.arange(1000000)

# Список (медленно)
%timeit [x**2 for x in lst]

# NumPy (быстро)
%timeit arr**2
```
